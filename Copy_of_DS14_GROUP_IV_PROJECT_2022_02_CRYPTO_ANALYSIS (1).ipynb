From the above info we'll change the Date column from object data type to date/time.

```python
#Changing the Date column's data type from object to date/time.
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])
bitcoin_data.info()
```

```python
#Changing the Date column's data type from object to date/time.
ethereum_data['Date'] = pd.to_datetime(ethereum_data['Date'])
ethereum_data.info()
```

```python
#Ensure uniformity by changing the column names to lower case, stripping of the whitespaces,removing periods, 
#this ensures good readerbility plus every naming issue is covered.
#This is to be applied to both the bitcoin and ethereum datasets.
#Bitcoin dataset

bitcoin_data.columns = bitcoin_data.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('.', '')
bitcoin_data.columns
```

```python
#Ethereum dataset
ethereum_data.columns = ethereum_data.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('.', '')
ethereum_data.columns
```

```python
#Renaming the columns in both our datasets to differentiate between the bitcoin and ethereum columns by adding an 'eth' suffix for ethereum
#and a 'bit' for bitcoin.
bitcoin_data2 = bitcoin_data.add_suffix('bit')
```

```python
#repeat the process for ethereum dataset

ethereum_data2 =ethereum_data.add_suffix('eth')
```

```python
bitcoin_data2.head()
```

```python
ethereum_data2.head()
```

```python
#renaming the 'dateeth' column to 'datebit' to be able to merge the two datasets since the data was collected in the same period.

ethereum_data3 = ethereum_data2.rename(columns={'dateeth': 'datebit'}, inplace=True)
```

```python
ethereum_data2
```

#Merging the datasets

```python
# creating a new data set by combining the Bitcoin and Ethereum datasets and displaying it.

crypto_data = pd.merge(bitcoin_data2, ethereum_data2, left_on='datebit', right_on='datebit')
crypto_data
```

```python
#Checking the info again
crypto_data.info()
```

After checking, there are no null values since the non-null count for each column = the total entries i.e 1554

```python
#Dropping the unnecessary columns 'closebit' and 'closeeth' since we can use 'adj_closebit' and 'adj_closeeth' which more conclusive figures.

crypto_data.drop(["closebit" , "closeeth"], axis = 1, inplace = True)
crypto_data.head(10)
```

```python
#Check Consistency and remove duplicated date records
crypto_data.drop_duplicates(subset=['datebit'])
```

```python
#Exporting our new clean and merged dataset to a single csv file.
crypto_data.to_csv("crypto_dataset.csv", index=False)
```

###5.What has been the percentage increase in price for the two currencies in comparison (Bitcoin and Etherium) over the past 5 years

```python
# Let's start by previewing the start and end highbit value of bitcoin in every year for the past 5 years.

range1=crypto_df[(crypto_df['datebit'] >= '2017-11-09') & (crypto_df['datebit'] <= '2018-11-09')].iloc[[0,-1]][['datebit','highbit']]
print(range1)
range2=crypto_df[(crypto_df['datebit'] > '2018-11-09') & (crypto_df['datebit'] <= '2019-11-09')].iloc[[0,-1]][['datebit','highbit']]
print(range2)
range3=crypto_df[(crypto_df['datebit'] > '2019-11-09') & (crypto_df['datebit'] <= '2020-11-09')].iloc[[0,-1]][['datebit','highbit']]
print(range3)
range4=crypto_df[(crypto_df['datebit'] > '2020-11-09') & (crypto_df['datebit'] <= '2021-11-09')].iloc[[0,-1]][['datebit','highbit']]
print(range4)
range5=crypto_df[(crypto_df['datebit'] > '2021-11-09') & (crypto_df['datebit'] <= '2022-11-21')].iloc[[0,-1]][['datebit','highbit']]
print(range5)
```

```python
from tabulate import tabulate
range1['Bitcoin Price Difference'] = range1['highbit'].diff(periods=1)

range2['Bitcoin Price Difference'] = range2['highbit'].diff(periods=1)

range3['Bitcoin Price Difference'] = range3['highbit'].diff(periods=1)

range4['Bitcoin Price Difference'] = range4['highbit'].diff(periods=1)

range5['Bitcoin Price Difference'] = range5['highbit'].diff(periods=1)

range1['Bitcoin Price % Profit/Loss'] = ((range1['Bitcoin Price Difference']/7446.83)*100)

range2['Bitcoin Price % Profit/Loss'] = ((range2['Bitcoin Price Difference']/6437.28)*100)

range3['Bitcoin Price % Profit/Loss'] = ((range3['Bitcoin Price Difference']/9103.83)*100)

range4['Bitcoin Price % Profit/Loss'] = ((range4['Bitcoin Price Difference']/15450.3)*100)

range5['Bitcoin Price % Profit/Loss'] = ((range5['Bitcoin Price Difference']/68789.63)*100)

print(tabulate(range1, headers='keys', tablefmt='psql'))
print(tabulate(range2, headers='keys', tablefmt='psql'))
print(tabulate(range3, headers='keys', tablefmt='psql'))
print(tabulate(range4, headers='keys', tablefmt='psql'))
print(tabulate(range5, headers='keys', tablefmt='psql'))
```

As per the data above, all the negative values indicate a price drop, hence a loss. The positive values, on the other hand, indicate a price increase, hence a profit. This analysis is done for the opening value and closing value of Bitcoin for every year for the period of five years.

From the Analysis above, Bitcoin experience the highest profit of 344% between 10-11-2020 and 09-11-21. It can also been concluded that Bitcoin experince the greatest loss between 10-11-2021 and 21-02-2022, which translated to 43% loss.
